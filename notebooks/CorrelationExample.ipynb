# Lagged correlation example 

```python
sigma = 0.1
lag = 0.5
x = np.arange(20, step=0.1)
y1 = np.cos(x) + sigma * np.random.standard_normal(len(x))
y2 = np.cos(x + lag) + sigma * np.random.standard_normal(len(x))
```

```python
plt.plot(y1)
plt.plot(y2)
```
